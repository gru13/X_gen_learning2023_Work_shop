```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load the dataset using Pandas
diabetes_df = pd.read_csv('diabetes.csv')

# Select the Age and Glucose columns
X = diabetes_df[['Age']]
y = diabetes_df['Glucose']

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Get the intercept and slope of the linear regression line
intercept = model.intercept_
slope = model.coef_[0]

# Print the equation of the linear regression line
print(f"Glucose = {intercept:.2f} + {slope:.2f}*Age")
print(intercept+slope*70)

# Plot the data points and linear regression line
plt.scatter(X, y, alpha=0.5)
plt.plot(X, model.predict(X), color='red')
plt.title('Age vs Glucose')
plt.xlabel('Age')
plt.ylabel('Glucose')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Load the dataset using Pandas
diabetes_df = pd.read_csv('diabetes.csv')

# Replace 0 values with NaNs in relevant columns
diabetes_df[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']] = diabetes_df[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']].replace(0, np.NaN)

# Remove rows with missing values
diabetes_df.dropna(inplace=True)

# Define X and y
X = diabetes_df[['Glucose']]
y = diabetes_df['Outcome']

# Train Logistic Regression model
log_reg = LogisticRegression()
log_reg.fit(X, y)

# Create a meshgrid of points to plot the sigmoid curve
x_min, x_max = X.min(), X.max()
xx = np.linspace(x_min, x_max, len(X)).reshape(-1, 1)
yy = log_reg.predict_proba(xx)[:,1]


# Plot the sigmoid curve
plt.plot(xx, yy)
plt.xlabel('Glucose')
plt.ylabel('Probability of Diabetes')
plt.show()

b0 = log_reg.intercept_[0]
b1 = log_reg.coef_[0][0]
print(b0,b1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

# Load data from CSV file
data = pd.read_csv("diabetes.csv")

# Separate features and target variable
X = data.iloc[:, [0, 1]].values
y = data.iloc[:, -1].values

# Create a meshgrid to plot the decision boundary
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))

# Create a KNN classifier with k=3
classifier = KNeighborsClassifier(n_neighbors=3)

# Train the classifier on the data
classifier.fit(X, y)

# Predict the class labels for the meshgrid points
Z = classifier.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundary and data points
plt.contourf(xx, yy, Z, alpha=0.4)
plt.scatter(X[:, 0], X[:, 1], c=y, alpha=0.8)
plt.xlabel('Pregnancies')
plt.ylabel('Glucose')
plt.title('KNN decision boundary')
plt.show()
```
